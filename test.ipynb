# Research Question

*Does Syria's outward rate of migration have an impact on where disturbed populations seek asylum and ultimately resettle?*

## Background and Prior Work


In the modern socio-political climate Refugee and Asylum seekers comprise one of the most vulnerable and one of the fastest growing populations. A 2021 study of refugee and asylum seekers by the United Nations indicated that a record number 84 million refugees had been displaced in the time period between January 2021 and November 2021 [^UN2]. Current projections show the effects of climate change, the COVID-19 Pandemic and governmental instability exacerbating this already growing trend throughout the 2020s. Understanding refugee patterns and the taken towards resettlement is key to being properly prepared and integrating them into host countries. Representing the Largest and longest refugee crisis of the 2010s, an analysis of refugee data from the Syrian Civil War can provide insights on how governments and NGOs can better prepare for future global crises.

Following the most intense drought in modern Syrian history from 2006 to 2010, the largely agrarian population had resettled into urban ghettos[^Brit]. Following the period of resettlement, disturbed populations began to protest inequitable economic policy as well as the complacent Assad regime[^Brit]. Military response to the peaceful protests seen throughout the region sparked what has become a decade long civil war for control of Syria. The Civil War has seen over 6.6 million registered Syrian refugees leave the country [^UN].
